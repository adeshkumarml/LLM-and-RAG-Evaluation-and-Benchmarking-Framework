```python
def chunk_text(text, chunk_size = 100):
    words = text.split()
    chunks = []

    for i in range(0, len(words), chunk_size):
        chunk = words[i: i+chunk_size]
        chunks.append(" ".join(chunk))

    return chunks
```

```python
import json

with open("data/dataset.json", "r") as f:
    processed_data = json.load(f)

print(len(processed_data))
```

```python
sample_chunks = chunk_text(sample_context)
print(len(sample_chunks))
print(sample_chunks[0])
```

```python
# chunked corpus with metadata
chunked_corpus = []
chunk_id = 0

for item in processed_data:
    context = item["context"]
    chunks = chunk_text(context)

    for chunk in chunks:
        chunked_corpus.append({
            "chunk_id": chunk_id,
            "source_id": item["id"],
            "text": chunk
        })

        chunk_id += 1
```

```python
# embedding
from sentence_transformers import SentenceTransformer
```

```python
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
chunk_texts = [chunk["text"] for chunk in chunked_corpus]

chunk_embeddings = embedding_model.encode(chunk_texts, show_progress_bar = True)
```

```python
print(type(chunked_corpus))
print(type(chunk_embeddings))
print(chunk_embeddings.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
# retrieval-engine function
def retrieve_top_k(query, k = 3):
    query_embedding = embedding_model.encode([query])
    
    similarities = cosine_similarity(query_embedding, chunk_embeddings)[0]
    top_k_idx = np.argsort(similarities)[-k:][::-1]
    results = [chunked_corpus[idx] for idx in top_k_idx]

    return results
```

```python
sample_query = processed_data[0]["question"]
print("Query: ",sample_query)
results = retrieve_top_k(sample_query)
print("Top retrieved chunks:")

for idx, result in enumerate(results, 1):
    print(f"Rank {idx} - {result['text']}\n")
```
